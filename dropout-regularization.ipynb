```python
x = df.drop(60, axis='columns')
y= df[60]
```

```python
# One hot encoding for metal and rock and droping the first column
y = pd.get_dummies(data=y,drop_first=True )  
y.head(5)  ## R->True   M->Flase
```

```python
y = y.replace({True:1, False: 0})
y.head(5)
```

```python
from sklearn.model_selection import train_test_split
x_trin, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=1)
```

```python
x_trin.shape, x_test.shape
```

```python
import tensorflow as tf
from tensorflow import keras 
```

```python
model = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(x_trin, y_train, epochs=100, batch_size=8)
```

```python
model.evaluate(x_test, y_test)
```

```python
y_pred = model.predict(x_test)
print(y_pred[:10])

y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
print(y_test[:10])
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

### Drop out Regularization
Mostly used in the computer vision related problems.

```python
modeld = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'),
    keras.layers.Dropout(0.5),  ## Drop neurons by 50%
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dropout(0.3),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dropout(0.8),
    keras.layers.Dense(1, activation='sigmoid')
])

modeld.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
modeld.fit(x_trin, y_train, epochs=100, batch_size=8)
```

```python
modeld.evaluate(x_test, y_test)
```

```python
y_pred = modeld.predict(x_test)
y_pred = np.round(y_pred)

from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
