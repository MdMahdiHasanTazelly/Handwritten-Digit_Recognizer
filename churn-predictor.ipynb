```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
df = pd.read_csv("./data/customer-churn.csv")
df
```

```python
df.drop('customerID', axis='columns', inplace=True)
df
```

```python
df.dtypes
```

```python
df.TotalCharges.values
```

```python
#shows how many spaces are in TotalCharges column
df[df.TotalCharges==' ']
```

```python
#new dataframes without spaces
new_df = df[df.TotalCharges!=' ']
```

```python
new_df[new_df.TotalCharges==' ']
```

```python
# converting to numeric type
new_df['TotalCharges'] = pd.to_numeric(new_df.TotalCharges)
```

```python
new_df['TotalCharges'].values
```

### Visualization

```python
tenure_churn_no = new_df[new_df.Churn=='No'].tenure
tenure_churn_yes = new_df[new_df.Churn=='Yes'].tenure

plt.figure(figsize=(5, 3))
plt.xlabel('Tenure')
plt.ylabel('No of Customers')
plt.title('Customer Churn Based on Tenure')
plt.hist([tenure_churn_yes, tenure_churn_no], color=['green', 'red'], label=['churn yes', 'churn no'])
plt.legend()
```

```python
mc_churn_no = df[df.Churn=='No'].MonthlyCharges
mc_churn_yes = df[df.Churn=="Yes"].MonthlyCharges

plt.figure(figsize=(5,3))
plt.title("Customer Churn Based on Monthly Charges")
plt.xlabel('Monthly charges')
plt.ylabel('No of Customers')
plt.hist([mc_churn_yes, mc_churn_no], color=['green', 'red'], label=['churn yes', 'churn no'])
plt.legend()
```

```python
def unique_col_values(df):
    for column in df:
        # if df[column].dtype=='object':
            print(f"{column} : {df[column].unique()}")
```

```python
unique_col_values(new_df)
```

```python
new_df.replace('No internet service', 'No', inplace=True)  
new_df.replace('No phone service', 'No', inplace=True)
```

```python
unique_col_values(new_df)
```

```python
#replacing all the yes by 1 and no by 0
columns_to_replace = ['Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
                      'TechSupport', 'StreamingTV', 'StreamingMovies', 'PaperlessBilling', 'Churn']

new_df[columns_to_replace] = new_df[columns_to_replace].replace({"Yes":1, "No":0})
```

```python
unique_col_values(new_df)
```

```python
#replacing male with 0 and female with 1
new_df['gender'].replace({'Male':0, 'Female':1}, inplace=True)
new_df['gender'].unique()
```

```python
#One hot encoding for InternetService, Contract, PaymentMethod
new_df_2 = pd.get_dummies(data=new_df, columns=['InternetService', 'Contract', 'PaymentMethod'])
new_df_2.columns
```

```python
new_df_2.sample(5)
```

```python
new_df_2.dtypes
```

```python
bool_columns= ['InternetService_DSL', 'InternetService_Fiber optic', 'InternetService_No', 'Contract_Month-to-month',
               'Contract_One year', 'Contract_Two year', 'PaymentMethod_Bank transfer (automatic)',
               'PaymentMethod_Credit card (automatic)', 'PaymentMethod_Electronic check', 'PaymentMethod_Mailed check']

new_df_2[bool_columns] = new_df_2[bool_columns].replace({True:1, False:0})
```

```python
new_df_2.dtypes
```

```python
new_df_2.sample(5)
```

### Data Scaling

```python
#converts these columns value within 0-1 range
cols_to_scale = ['tenure', 'MonthlyCharges', 'TotalCharges']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
new_df_2[cols_to_scale] = scaler.fit_transform(new_df_2[cols_to_scale])
new_df_2.sample(3)
```

### Train Test Splitting

```python
x = new_df_2.drop('Churn', axis='columns')
y = new_df_2['Churn']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=5)
```

```python
len(x_train.columns)
```

### Model Training

```python
import tensorflow as tf
from tensorflow import keras 

model = keras.Sequential([
    keras.layers.Dense(20, input_shape=(26, ), activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid'),
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.fit(x_train, y_train, epochs=150)
```

### Evaluating

```python
model.evaluate(x_test, y_test)
```

```python
yp = model.predict(x_test)
yp[:10]
```

```python
y_test[:10]
```

```python
y_pred = []
for elem in yp:
    if elem>0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)    

y_pred[:10]
```

### Classification Report

```python
from sklearn.metrics import confusion_matrix, classification_report
print(classification_report(y_test, y_pred))
```

```python
#Confusion Metrics
cm = tf.math.confusion_matrix(labels=y_test, predictions=y_pred)

plt.figure(figsize=(5, 3))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
